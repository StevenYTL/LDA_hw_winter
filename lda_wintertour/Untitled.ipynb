```python
os.chdir("C:\\Users\\steven\\Desktop\\lda_wintertour")

stopWords=[]

with open('stopWords.txt', 'r',encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopWords.append(data)
        
with open('stopWords2.txt', 'r',encoding='UTF-8') as file2:
    for data in file2.readlines():
        data = data.strip()
        stopWords.append(data)
        
with open('stopWords3.txt', 'r',encoding='UTF-8') as file3:
    for data in file3.readlines():
        data = data.strip()
        stopWords.append(data)
        
with open('stopWords4.txt', 'r',encoding='UTF-8') as file4:
    for data in file4.readlines():
        data = data.strip()
        stopWords.append(data)        

print(len(stopWords))
```

```python
remainderWords=[]
#重要filter
remainderWords = list(filter(lambda a: a not in stopWords and a != '\n' and a != ' ', words)) 

print(len(remainderWords))
```

```python
with open('corpus_stop.txt','w',newline='',encoding='utf-8-sig') as article:
    for item in remainderWords:
        article.write(item)
        article.write(" ")
```
